```python
# parents_classes = ['bachelor', 'masters', 'college', 'masters', 'some high', 'assosiate']
# race_classes = ['A', 'B', 'C', 'D', 'E']
gender_classes = {'male' : 0, 'female' : 1}
lunch_classes = {'free/reduced' : 0, 'standard' : 1}
preparation_classes = {'none' : 0, 'completed' : 1}
parents_classes = {"bachelor's degree" : 1, "master's degree" : 1, "some college" : 1, "associate's degree" : 1, "some high school" : 0, "high school" : 0}
race_classes = {'group E' : 1, 'group D' : 1, 'group A' : 0, 'group B' : 0, 'group C' : 0}

# df[parents_classes] = pd.get_dummies(df['parental level of education'], columns = parents_classes)
df['parental level of education'] = df['parental level of education'].map(parents_classes)
# df[race_classes] = pd.get_dummies(df['race/ethnicity'], columns = race_classes)
df['race/ethnicity'] = df['race/ethnicity'].map(race_classes)
df['gender'] = df['gender'].map(gender_classes)
df['lunch'] = df['lunch'].map(lunch_classes)
df['test preparation course'] = df['test preparation course'].map(preparation_classes)
```

```python
sns.barplot(x = 'race/ethnicity', y = 'total_points', data = df)
print(df[df['race/ethnicity'] == 0]['total_points'].mean())
print(df[df['race/ethnicity'] == 1]['total_points'].mean())
```

```python
print(df[df['parental level of education'] == 0]['total_points'].mean())
print(df[df['parental level of education'] == 1]['total_points'].mean())
sns.barplot(x = 'parental level of education', y = 'total_points', data = df)
```

```python
df2 = df.drop(['math score', 'reading score', 'writing score', 'total_points'], axis = 1)   
```

```python
y = df['total_points']
```

```python
df2.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df2, y, test_size = 0.2, random_state = 1)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
model.coef_
```

```python
df.head()
```

```python
plt.scatter(y_test, prediction)
```

```python
from sklearn import metrics

print('MEA:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
