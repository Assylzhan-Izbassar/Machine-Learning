## 1. Fetching the dataset

```python
import os
import tarfile
from six.moves import urllib
```

```python
download_root = 'https://raw.githubusercontent.com/ageron/handson-ml/master/'
housing_path = os.path.join('datasets', 'housing')
housing_url = download_root + 'datasets/housing/housing.tgz'
```

```python
def fetch_housing_data(hsing_url = housing_url, hsing_path = housing_path):
    if not os.path.isdir(hsing_path):
        os.makedirs(hsing_path)
    tgz_path = os.path.join(hsing_path, 'housing.tgz')
    urllib.request.urlretrieve(hsing_url, tgz_path)
    hsing_tgz = tarfile.open(tgz_path)
    hsing_tgz.extractall(path=hsing_path)
    hsing_tgz.close()
```

```python
def load_housing_data(hsing_path=housing_path):
    csv_path = os.path.join(hsing_path, 'housing.csv')
    return pd.read_csv(csv_path)
```

## 2. Create a Test Set

```python
np.random.seed(69)
```

```python
# Simple split the data to train and test
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(hsing, 0.2)
print(len(train_set), 'train +', len(test_set), 'test')
```

```python
import hashlib
```

```python
# test set never contain the data of train set in multiple fetching
def test_set_check(data_id, test_ratio, hash):
    return hash(np.int64(data_id)).digest()[-1] < 256 * test_ratio
```

```python
def split_train_test_by_id(data, test_ratio, id_column, hash=hashlib.md5):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio, hash))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
hsing_with_id = hsing.reset_index() # adds an 'index' column
train_set, test_set = split_train_test_by_id(hsing_with_id, 0.2, 'index')
```

One issue of this new fetching data have to be append and the data have not delete, otherwise it is not work properly.

```python
# To overcome this issue we can use 'longitude' and 'latitude' as id
hsing_with_id['id'] = hsing['longitude']*1000+hsing['latitude']
train_set, test_set = split_train_test_by_id(hsing_with_id, 0.2, 'id')
```

```python
hsing_with_id.head()
```

Checking that id is not repeated:

```python
test_id_vals = np.array(test_set.id)
train_id_vals = np.array(train_set.id)
```

```python
cnt = 0
for i in range(0, len(train_id_vals)):
    if(test_id_vals[i%len(test_id_vals)] == train_id_vals[i]):
        cnt = cnt+1
print(cnt)
```

### Stratified sampling based on the income category

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size = 0.2, random_state = 69)
for train_idx, test_idx in split.split(hsing, hsing['income_category']):
    strat_train_set = hsing.loc[train_idx]
    strat_test_set = hsing.loc[test_idx]
```
